```python
import pandas as pd
import numpy as np
import matplotlib as pt
import os
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = 20, 8
import plotly.graph_objs as go
import chart_studio.plotly
```

```python
df=pd.read_csv("final.csv")
df
```

```python
df['State_Mean'] = df.groupby(['State']).Tons.transform('mean')
df['Market_Mean'] = df.groupby(['Market']).Tons.transform('mean')
df['State_Price_Mean'] = df.groupby(['State']).Modal_Price.transform('mean')
df['Market_Price_Mean'] = df.groupby(['Market']).Modal_Price.transform('mean')
```

```python
State = sns.countplot(y='State', data=df)
```

```python
State = sns.countplot(y='Month', data=df)
```
