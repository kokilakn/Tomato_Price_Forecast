```python
import pandas as pd
import numpy as np
import matplotlib as pt
import os
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = 20, 8
import plotly.graph_objs as go
import chart_studio.plotly
```

```python
df=pd.read_csv("final.csv")
```

```python
# filling the missing data with appropriate data
```

```python
df = df[df['Max_Price'].notna()]
df['Modal_Price'] = df['Modal_Price'].fillna((df['Max_Price']+df['Min_Price'])/2)
```

```python
df.isnull().sum()
```

```python
#filling the zero values
```

```python
(df == 0).sum()
```

```python
df['Modal_Price'] = np.where(((df['Modal_Price'] == 0) & (df['Max_Price'] != 0) & (df['Min_Price'] != 0 ))
                              , ((df['Max_Price']+df['Min_Price'])/2), df['Modal_Price'])
```

```python
df = df[df.Modal_Price != 0]
```

```python
df['Max_Price'] = np.where(((df['Max_Price'] == 0) & (df['Min_Price'] != 0))
                              , (((df['Modal_Price'])*2)-df['Min_Price']), df['Max_Price'])
df = df[df.Max_Price != 0]
```

```python
df['Min_Price'] = np.where(((df['Min_Price'] == 0) & (df['Max_Price'] != 0))
                              , (((df['Modal_Price'])*2)-df['Max_Price']), df['Min_Price'])
df = df[df.Min_Price != 0]
```

```python
df = df[df.Tons != 0]
(df == 0).sum()
```

```python
plt.figure(figsize=(16,7))
Variety = sns.countplot(x='Variety', data=df)
```

```python
#as local and desi is same and few markets consider tomato as local and sankar is a hybrid where is considered to be local
df['Variety'] = np.where(df['Variety'] == 'Deshi'
                              , 'Local', df['Variety'])
df['Variety'] = np.where(df['Variety'] == 'Tomato'
                              , 'Local', df['Variety'])
df['Variety'] = np.where(df['Variety'] == 'Other'
                              , 'Local', df['Variety'])
df['Variety'] = np.where(df['Variety'] == 'Sankar'
                              , 'Hybrid', df['Variety'])
```

```python
plt.figure(figsize=(16,7))
Variety = sns.countplot(x='Variety', data=df)
```

```python
#saving the edited data
df.to_csv('final.csv',index=False)
```
