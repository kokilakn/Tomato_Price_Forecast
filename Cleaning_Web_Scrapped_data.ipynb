```python
import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
header_list = ["State", "District", "Market","Variety","Commadity","Tons","Min_Price","Max_Price","Modal_Price","Arrival_Date"]
```

```python
df1=pd.read_csv("data1.csv",encoding='latin1', names=header_list, decimal=',')
df2=pd.read_csv("data2.csv",encoding='latin1', names=header_list, decimal=',')
df3=pd.read_csv("data3.csv",encoding='latin1', names=header_list, decimal=',')
```

```python
out=df1.append(df2).append(df3)
out
```

```python
df=out
df = df[df.State != "-"]
```

```python
df
```

```python
df['Arrival_Date'] = pd.to_datetime(df['Arrival_Date'], format='%d %b %Y')
```

```python
df.isnull().sum()
```

```python
df[df['Arrival_Date'].isnull()]
```

```python
# removing the NaN data
df = df[df['Arrival_Date'].notna()]
```

```python
df['Year'] = pd.DatetimeIndex(df['Arrival_Date']).year
df['Month'] = pd.DatetimeIndex(df['Arrival_Date']).month
df['Date'] = pd.DatetimeIndex(df['Arrival_Date']).day
```

```python
#sorting data by date
df = df.sort_values("Arrival_Date")
df
```

```python
df.info()
```

```python
#removing unwanted data
df['Max_Price']=df['Max_Price'].replace(r'^\s*$', np.nan, regex=True)
df['Modal_Price']=df['Modal_Price'].replace(r'^\s*$', np.nan, regex=True)
df['Modal_Price']=df['Modal_Price'].replace(r'Â\xa0', np.nan, regex=True)
```

```python
#changing the type of columns
df["Date"]=df.Date.astype(int)
df["Month"]=df.Month.astype(int)
df["Year"]=df.Year.astype(int)
df["Min_Price"]=df.Min_Price.astype(float)
df["Max_Price"]=df.Max_Price.astype(float)
df["Modal_Price"]=df.Modal_Price.astype(float)
df['Tons'] = df['Tons'].str.replace(',', '')
df["Tons"]=df.Tons.astype(float)
```

```python
# exporting the edit df to final csv file
df.to_csv('final.csv',index=False)
```
